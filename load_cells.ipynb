```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
cm = 1/2.54  # centimeters in inches
fig, axs = plt.subplots(figsize=(10*cm,8*cm))
```

```python
# Cells

alpha = 0.7
x = np.linspace(0, 110)
y1 = 3000*np.ones_like(x)
y2 = 27000*np.ones_like(x)
plt.fill_between(x, y1, y2, alpha = alpha)


x = np.linspace(0, 100)
y1 = 1000*np.ones_like(x)
y2 = 9000*np.ones_like(x)
plt.fill_between(x, y1, y2, alpha = alpha)

x = np.linspace(0, 90)
y1 = 500*np.ones_like(x)
y2 = 4500*np.ones_like(x)
plt.fill_between(x, y1, y2, alpha = alpha)

x = np.linspace(0, 80)
y1 = 200*np.ones_like(x)
y2 = 1800*np.ones_like(x)
plt.fill_between(x, y1, y2, alpha = 0.2)


x = np.linspace(0, 70)
y1 = 100*np.ones_like(x)
y2 = 900*np.ones_like(x)
plt.fill_between(x, y1, y2, alpha = alpha)

x = np.linspace(0, 60)
y1 = 50*np.ones_like(x)
y2 = 450*np.ones_like(x)
plt.fill_between(x, y1, y2, alpha = alpha)


# specimens
type = [1,2] # ASTM D638
width = [13,6] # mm
thickness = [3,5] # mm

sigma = np.logspace(0,2) # tensile stress

for b in width:
    for d in thickness:
        F = sigma * b * d 
        
        if b == width[0]:
            line = '-'
            label1 = 'Type 1'
        else:
            line = '--'
            label1 = 'Type 2'
        
        if d == thickness[0]:
            color = 'k'
            label2 = f' , thickness = {thickness[0]} mm'
        else:
            color = 'b' 
            label2 = f' , thickness = {thickness[1]} mm'
            
        label = label1 + label2
        plt.loglog(sigma,F, ls = line, color = color, label = label)
        
    

plt.ylabel('Force (N) ')
plt.xlabel('Stress (MPa)')
plt.legend()

#plt.ylim(bottom=10, top=100000)
plt.legend()
#plt.legend(fontsize='small')


plt.savefig(f'load_cells.png', dpi=600)
```
